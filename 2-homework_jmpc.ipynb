```python

#load data 
# Create a range of values from -10 to 10
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)

# Create a grid of points
X, Y = np.meshgrid(x, y)

# Define the Sigma matrix, is positive definite and symetric
#Sigma=np.array([[50, 30], [30, 50]])

#Sigma = np.array([[2, 1], [1, 3]])
#Sigma = np.array([[0.5, 0.3], [0.3, 0.5]])
Sigma = np.array([[0.05, 0.03], [0.03, 0.05]])

#


print(np.linalg.eigvals(Sigma))
```

```python

# compute de values of the norm .

if is_positive_definite(Sigma) == True:
    Z = np.empty(X.shape)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            vec = np.array([X[i, j], Y[i, j]])
            Z[i, j] = compute_norm_excercise(vec, Sigma)

else:
    print("The matrix is not positive definite and symetric")

```

```python
plot_norm_excersice(X, Y, Z)
```
