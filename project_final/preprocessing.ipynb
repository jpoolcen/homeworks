# Preprocessing Dataset Gesture Phase 

```python
import pandas as pd
import numpy as np
import glob
# source https://archive.ics.uci.edu/dataset/302/gesture+phase+segmentation
```

```python
base_path = 'homeworks/data/project/gesture_phase/'
```

```python
files = glob.glob(base_path + '*_va3.csv')
```

```python
dfs=[]
for file in files:
    df = pd.read_csv(file)
    dfs.append(df)  
df_all = pd.concat(dfs, ignore_index=True)    
```

```python
print(df_all.shape)
```

```python
df_all.head()
```

```python
df_all.to_csv(base_path + 'gesture_phase.csv', index=False)
```

```python
base_path = 'homeworks/data/project/'
```

```python
df_shruntime = pd.read_csv(base_path + 'shrutime.csv')
```

```python
df_shruntime.head()
```

```python
df_encoded = pd.get_dummies(df_shruntime, columns=['Geography', 'Gender'])

df_encoded.head()

print(df_encoded.columns)
```

```python
df_filter_encoded = df_encoded[['CreditScore', 'Age', 'Tenure',
       'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember',
       'EstimatedSalary', 'Geography_France', 'Geography_Germany',
       'Geography_Spain', 'Gender_Female', 'Gender_Male','Exited']]
```

```python
df_filter_encoded.head()
```

```python
## standarize the data

from sklearn.preprocessing import StandardScaler

# Crear el estandarizador
scaler = StandardScaler()

# Seleccionar las columnas a estandarizar
columns_to_standardize = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
features_to_standardize = df_filter_encoded[columns_to_standardize]

# Ajustar el estandarizador a los datos y transformarlos
standardized_features = pd.DataFrame(scaler.fit_transform(features_to_standardize), columns=features_to_standardize.columns)


```

```python
standardized_features
```

```python
dummies_and_label = df_filter_encoded[['HasCrCard', 'IsActiveMember', 'Geography_France', 'Geography_Germany', 'Geography_Spain', 'Gender_Female', 'Gender_Male', 'Exited']]

# Concatenar standardized_features con las columnas dummies y la columna de salida
df_final = pd.concat([standardized_features, dummies_and_label], axis=1)

df_final.head()
```

```python
df_final.to_csv(base_path + 'shrutime_encoded.csv', index=False)
```
